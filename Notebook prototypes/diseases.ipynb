# Filtering data

```python
print(df["response"].unique())
print(df["response_id"].unique())

print(df["stratification_2"].unique())
print(df["stratification_id_2"].unique())
print(df["stratification_cat_2"].unique())
print(df["stratification_cat_id_2"].unique())

print(df["stratification_3"].unique())
print(df["stratification_id_3"].unique())
print(df["stratification_cat_3"].unique())
print(df["stratification_cat_id_3"].unique())
```

```python
filtered_df = df.drop(columns = ["response","response_id","stratification_2","stratification_id_2","stratification_cat_2","stratification_cat_id_2","stratification_3","stratification_id_3","stratification_cat_3","stratification_cat_id_3"])
```

```python
#Some of the location data won't be necessary
filtered_df = filtered_df.drop(columns = ["geolocation", "location_id"])
```

```python
# filtered_df should have 6 columns less than the original one.
filtered_df
```

```python
filtered_df = filtered_df[filtered_df["data_value_footnote"].isna()]
```

```python
#After removing data that might be trusted, the df should have a little bit more than 200,000 rows
len(filtered_df)
```

```python
#Now, the data_value column must only be considered on rows without NaNs
filtered_df = filtered_df[filtered_df["data_value"].notna()]
```

```python
#Once again, the rows of the df are checked
len(filtered_df)
```

```python
#Since the "data_value_footnote" column won't be used anymore, it's possible to drop it and save some storage.
filtered_df = filtered_df.drop(columns = ["data_value_footnote", "data_value_footnote_symbol"])
```

```python
filtered_df.columns
```

```python
#The following filter makes sure there won't be missing States in the data.
filtered_df = filtered_df[filtered_df["state_abr"].notna()]
filtered_df = filtered_df[filtered_df["state_abr"] != None]

filtered_df = filtered_df[filtered_df["state_name"].notna()]
filtered_df = filtered_df[filtered_df["state_name"] != None]
```
